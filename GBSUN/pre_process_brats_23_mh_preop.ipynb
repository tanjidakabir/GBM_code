```python
def savepkl(data,path):
    with open(path, 'wb') as f:
        pickle.dump(data, f)
```

```python
def image_process(data_index):
    t1_path=data_index.t1
    t1ce_path=data_index.t1ce
    t2_path=data_index.t2
    flair_path=data_index.flair
    seg_path=data_index.seg
    
    img_t1 = nib.load(t1_path).get_fdata()
    img_t1=np.array(img_t1,dtype='float32', order='C')
    
    img_t1ce = nib.load(t1ce_path).get_fdata()
    img_t1ce=np.array(img_t1ce,dtype='float32', order='C')
    
    img_t2 = nib.load(t2_path).get_fdata()
    img_t2 =  np.array(img_t2,dtype='float32', order='C')
    
    img_flair = nib.load(flair_path).get_fdata()
    img_flair=np.array(img_flair,dtype='float32', order='C')
    
    images=np.stack([img_t1,img_t1ce,img_t2,img_flair ],-1)
    
    label = nib.load(seg_path).get_fdata().astype('long')
    
    
    label[label == 1] = 5
    label[label == 2] = 6
    label[label == 3] = 7
    
    label[label == 5] = 3
    label[label == 6] = 1
    label[label == 7] = 2
    
    #label[label == 1] = 4
    #label[label == 2] = 2
    #label[label == 3] = 1
    #label[label == 4] = 3
   
    
    #label[label == 1] = 5
    #label[label == 2] = 6
    #label[label == 4] = 7
    #label[label == 5] = 2
    #label[label == 6] = 1
    #label[label == 7] = 1"
    
    
    #label[label == 1] = 1
    #label[label == 2] = 1
    #label[label == 4] = 0
    #label[label == 3] = 5
    #label[label == 5] = 2



    
    
    #label[label > 1] = 0  # Replace label 4 with label 0
    #label = np.array(label, dtype='uint8', order='C')
    mask = images.sum(-1) > 0
    for k in range(4):
        x = images[...,k] #
        y = x[mask] #
        
        lower = np.percentile(y, 0.2) # 
        upper = np.percentile(y, 99.8)
        
        x[mask & (x < lower)] = lower
        x[mask & (x > upper)] = upper

        y = x[mask]

        x -= y.mean()
        x /= y.std()

        images[...,k] = x
    
    
    
    images=images.transpose((3,0,1,2))
    
    return images, label
```

```python
process_images_path='/home/Kang_lin_post_model/pre_process_data_brats_mh_preop/images_23/'
process_seg_path='/home/Kang_lin_post_model/pre_process_data_brats_mh_preop/segmentation_23/'
```

```python
all_images_path=[]
all_seg_path=[]
all_id=[]
for i in tqdm(range(len(train_df))):
    data_index=train_df.iloc[i]
    images, label=image_process(data_index)
    images_name=data_index.pt_id+'_images.pkl'
    images_path=process_images_path+images_name
    all_images_path.append(images_path)
    seg_name=data_index.pt_id+'_seg.pkl'
    seg_path=process_seg_path+seg_name
    all_seg_path.append(seg_path)
    all_id.append(data_index.pt_id)
    print("saving:",data_index.pt_id+'_information')
    savepkl(images,images_path)
    savepkl(label,seg_path)
returndf=pd.DataFrame(all_id,columns=['pt_id'])
returndf['images_path']=all_images_path
returndf['seg_path']=all_seg_path
returndf.to_csv('/home/Kang_lin_post_model/pre_process_data_brats_mh_preop/BRATS_2023_preop_index_2.csv',index=False)
```
